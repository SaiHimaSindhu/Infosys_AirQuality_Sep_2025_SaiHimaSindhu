```python
import plotly.express as px
import pandas as pd
```

```python
# Load your dataset
df = pd.read_csv("air_quality.csv", parse_dates=['Date'])
pollutants = ['PM2.5','PM10','NO2','SO2','CO']
```

```python
# 1. Line Plot
px.line(df, x='Date', y='PM2.5', title='PM2.5 Over Time').show()
```

```python
# 2. Scatter Plot
px.scatter(df, x='PM2.5', y='PM10', color='NO2', title='PM2.5 vs PM10 vs NO2').show()
```

```python
# 3. Histogram
px.histogram(df, x='SO2', nbins=30, title='SO2 Histogram').show()
```

```python
# 4. Box Plot
px.box(df, y='PM10', title='Boxplot of PM10').show()
```

```python
# 5. Area Plot
px.area(df, x='Date', y='PM2.5', title='PM2.5 Area Plot').show()
```

```python
# 6. Correlation Heatmap
px.imshow(df[pollutants].corr(), text_auto=True, title='Correlation Heatmap').show()
```

```python
# 7. Pie Chart
avg = df[pollutants].mean().reset_index()
avg.columns = ['Pollutant', 'Mean']
px.pie(avg, names='Pollutant', values='Mean', title='Average Pollutant Share').show()
```

```python
# 8. Scatter Matrix
px.scatter_matrix(df[pollutants], title='Scatter Matrix of Pollutants').show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Create Month column
df['Month'] = df['Date'].dt.strftime('%b')

# Define pollutants
pollutants = ['PM10', 'PM2.5', 'SO2', 'NO2', 'CO', 'O3']

```

```python
# 9. Bar Chart
df_month = df.groupby('Month')[pollutants].mean().reset_index()
px.bar(df_month, x='Month', y='PM10', title='Monthly Average PM10').show()
```

```python
# 10. 3D Scatter
px.scatter_3d(df, x='PM2.5', y='PM10', z='NO2', color='SO2', title='3D Pollutant Relationship').show()
```

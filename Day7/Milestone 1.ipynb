```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df= pd.read_csv(r"AirQuality.csv", sep=';', low_memory=False)
```

```python
cleaned = df.drop(columns=[col for col in df.columns if col.startswith('Unnamed')])
```

```python
def clean_float(x):
    if pd.isnull(x):
        return np.nan
    try:
        return float(str(x).replace(',', '.'))
    except:
        return np.nan
```

```python
float_cols = ['CO(GT)','C6H6(GT)','T','RH','AH']
```

```python
for col in float_cols:
    cleaned[col] = cleaned[col].apply(clean_float)
```

```python
plt.figure(figsize=(12, 6))
```

```python
sns.histplot(cleaned['CO(GT)'].dropna(), bins=30, kde=True)
```

```python
plt.title('CO(GT) Distribution')
```

```python
plt.show()
```

```python
plt.figure(figsize=(12, 6))
```

```python
sns.lineplot(x=cleaned.index, y='CO(GT)', data=cleaned)
```

```python
plt.title('CO(GT) Time Series')
```

```python
plt.xlabel('Index (Time)')
```

```python
plt.ylabel('CO(GT)')
```

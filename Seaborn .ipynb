```python
import seaborn as sns
import pandas as pd
sns.set(style="whitegrid")
```

```python
# Load your dataset
df = pd.read_csv("air_quality.csv", parse_dates=['Date'])
pollutants = ['PM2.5','PM10','NO2','SO2','CO']
```

```python
# 1. Lineplot
sns.lineplot(x='Date', y='PM2.5', data=df).set_title('PM2.5 Over Time')
```

```python
# 2. Scatterplot
sns.scatterplot(x='PM2.5', y='PM10', data=df).set_title('PM2.5 vs PM10')
```

```python
# 3. Histogram + KDE
sns.histplot(df['NO2'], kde=True).set_title('NO2 Distribution')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.strftime('%b')

# Clean column names if necessary
df.rename(columns={'SO2 (µg/m³)': 'SO2'}, inplace=True)
```

```python
# 4. Boxplot by Month
sns.boxplot(x='Month', y='SO2', data=df).set_title('SO2 by Month')
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Create Month column
df['Month'] = df['Date'].dt.strftime('%b')

# Rename column if necessary
df.rename(columns={'PM2.5 (µg/m³)': 'PM2.5'}, inplace=True)

# Drop rows with missing values in these columns
df = df.dropna(subset=['Month', 'PM2.5'])
```

```python
# 5. Violinplot by Month
sns.violinplot(x='Month', y='PM2.5', data=df).set_title('PM2.5 by Month')
```

```python
# 6. Correlation Heatmap
sns.heatmap(df[pollutants].corr(), annot=True, cmap='coolwarm').set_title('Correlation Heatmap')
```

```python
# 7. Pairplot
sns.pairplot(df[pollutants])
```

```python
import matplotlib.pyplot as plt
```

```python
# 8. KDE Plot
for col in pollutants:
    sns.kdeplot(df[col], label=col)
plt.legend(); plt.title('Pollutant Distributions (KDE)')
```

```python
# 9. Stripplot
sns.stripplot(x='Month', y='PM10', data=df).set_title('PM10 by Month (Strip Plot)')
```

```python
# 10. Countplot
sns.countplot(x='Month', data=df).set_title('Data Count per Month')
```

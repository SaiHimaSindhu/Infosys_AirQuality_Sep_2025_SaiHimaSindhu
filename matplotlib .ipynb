```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load your dataset
df = pd.read_csv("air_quality.csv", parse_dates=['Date'])
pollutants = ['PM2.5','PM10','NO2','SO2','CO']
```

```python
# 1. Line Plot
plt.figure(figsize=(8,4))
plt.plot(df['Date'], df['PM2.5'])
plt.title('PM2.5 over Time')
plt.xlabel('Date'); plt.ylabel('PM2.5')
plt.show()
```

```python
# 2. Scatter Plot
plt.scatter(df['PM2.5'], df['PM10'])
plt.title('PM2.5 vs PM10'); plt.xlabel('PM2.5'); plt.ylabel('PM10')
plt.show()
```

```python
# 3. Histogram
plt.hist(df['NO2'], bins=30)
plt.title('Distribution of NO2'); plt.xlabel('NO2'); plt.ylabel('Frequency')
plt.show()
```

```python
# 4. Boxplot
plt.boxplot(df['PM10'].dropna())
plt.title('Boxplot of PM10'); plt.show()
```

```python
# 5. Bar Chart (Monthly Average)
df['Month'] = df['Date'].dt.month
plt.bar(df.groupby('Month')['SO2'].mean().index,
        df.groupby('Month')['SO2'].mean().values)
plt.title('Average SO2 by Month'); plt.xlabel('Month'); plt.ylabel('SO2')
plt.show()
```

```python
# 6. Correlation Heatmap
plt.imshow(df[pollutants].corr(), cmap='coolwarm', interpolation='nearest')
plt.xticks(range(len(pollutants)), pollutants)
plt.yticks(range(len(pollutants)), pollutants)
plt.title('Correlation Heatmap')
plt.colorbar()
plt.show()
```

```python
# 7. Area Plot
df.set_index('Date')[pollutants].plot.area(figsize=(10,4))
plt.title('Pollutant Trends (Area Plot)')
plt.show()
```

```python
# 8. Rolling Mean
plt.plot(df['Date'], df['PM2.5'].rolling(7).mean())
plt.title('7-Day Rolling Mean of PM2.5')
plt.show()
```

```python
# 9. Pie Chart
plt.pie(df[pollutants].mean(), labels=pollutants, autopct='%1.1f%%')
plt.title('Average Pollutant Contribution')
plt.show()
```

```python
# 10. Stacked Bar
df_month = df.groupby('Month')[pollutants].mean()
df_month.plot(kind='bar', stacked=True, figsize=(8,4))
plt.title('Stacked Bar of Monthly Pollutant Averages')
plt.show()
```
